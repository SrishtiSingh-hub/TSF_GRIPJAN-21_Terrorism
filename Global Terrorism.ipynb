```python
df.rename(columns={'eventid':'Event_Id','iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df.head(15)
```

```python
df['Calamity'] = df['Killed'] + df['Wounded']
df.head()
```

```python
display(df.isnull().sum().sort_values(ascending=False))
```

```python
Column = ['Event_Id','Year','Month','Day','Country','Region','AttackType','Target','Killed','Wounded','Summary','Group','Target_type','Weapon_type','Motive']
df_proc = df[~df.Killed.isna()][Column].reset_index(drop=True)
df_proc.head()
```

COUNTRIES THAT HAVE MORE VICTIMS

```python
df_proc.groupby(['Country'])['Killed'].sum().sort_values(ascending=False).head(7)
```

COUNTRIES THAT ARE MOSTLY AFFECTED

```python
sns.barplot(df_proc['Country'].value_counts()[:10].index,df_proc['Country'].value_counts()[:10].values)
plt.xlabel('Countries')
plt.ylabel('Number of attacks')
plt.xticks(rotation= 90)
plt.show()
```

HOT ZONE OF TERRIORSM

```python
def color_from_int(num,max_value):
    red = num/max_value
    blue = 1.0 - (num/max_value)
    green = 1.0 - (num/max_value)
    return (red, green, blue)
```

```python
df_temp = df.groupby(["Country"])["Event_Id"].count().reset_index(name="attack_count").sort_values("attack_count",ascending=False)
display(df_temp.head())
min_val = df_temp.loc[0:15]["attack_count"].min()
max_val = df_temp.loc[0:15]["attack_count"].max()
plt.figure(figsize=(10,10))
plt.title("HOT ZONE OF TERRIORSM")
plt.bar(x=df_temp[0:15]["Country"],height=df_temp[0:15]["attack_count"],align='center',color=df_temp.loc[0:15]["attack_count"].apply(lambda x, max_val=max_val: color_from_int(x,max_val)))
plt.xticks(rotation='vertical')
plt.show()
```

Number Of Terrorist Activities Each Year

```python
print("We have data for years {0} to {1}".format(df["Year"].min(),df["Year"].max()))
df_temp = df.groupby("Year")["Event_Id"].count().reset_index(name="attacks")
max_val = df_temp["attacks"].max()
display(df_temp)
```

Number Of Terrorist Activities Each Year

```python
plt.figure(figsize=(10,10))
plt.title("Number Of Terrorist Activities Each Year")
plt.xlabel("Years")
plt.ylabel("Number of attacks")
plt.plot(df_temp["Year"],df_temp["attacks"])
plt.show()
```

Attacks Distribution Of Regions By Year

```python
df_region=pd.crosstab(df.Year,df.Region)
df_region.plot(color=sns.color_palette('Set1',12))
fig=plt.gcf()
fig.set_size_inches(20,15)
plt.show()
```

Attack types and regions

```python
pd.crosstab(df.Region,df.AttackType).plot.barh(stacked=True,width=1,color=sns.color_palette('coolwarm',10))
fig=plt.gcf()
fig.set_size_inches(15,10)
plt.show()
```
